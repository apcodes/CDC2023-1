```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.arange(5)  # X-axis
y = np.arange(5)  # Y-axis
X, Y = np.meshgrid(x, y)  # Create a grid of X and Y values
Z = np.random.rand(5, 5) * 10  # Z-axis (height of bars)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.bar3d(X.ravel(), Y.ravel(), np.zeros_like(X).ravel(), 1, 1, Z.ravel(), shade=True)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
